1.	Data Preparation:

Load the dataset into pandas.

Convert the Date column to datetime format.

Extract year and month into separate columns.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('daily-min-temperatures.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
df.head()
```

2.	Analysis with Python:

	Create a Python function monthly_avg_temp(df, year) that returns the average temperature for each month in a specified year.

	Compute and display monthly average temperatures for the year 1988.



```python
def monthly_avg_temp(df, year):
    df_year = df[df['Date'].dt.year == year]
    monthly_avg = df_year.groupby(df_year['Date'].dt.month)['Temp'].mean()
    return monthly_avg
```

```python
avg_1988 = monthly_avg_temp(df, 1988)
print("Monthly Average Temperatures for 1988:\n")
print(avg_1988)
```

3.	Visualization:

Plot the daily temperatures for the entire dataset to visualize temperature trends over time.

Plot a line chart showing monthly average temperatures for the year 1988.


```python
plt.plot(df['Date'], df['Temp'])
plt.title('Daily Temperature Trend')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```

```python
avg_1988 = monthly_avg_temp(df, 1988)

plt.plot(avg_1988.index, avg_1988.values)
plt.title('Monthly Average Temperature - 1988')
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.show()
```
